```python
from transformers import BertTokenizer, BertModel
```

下载预训练的M-BERT模型以及它的词元分析器

```python
model = BertModel.from_pretrained('bert-base-multilingual-cased')
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
```

设定输入句子。这里使用一个法语句子作为输入。

```python
sentence = "C'est une si belle journée"
inputs = tokenizer(sentence, return_tensors="pt")
print(inputs)
```

```python
outputs = model(**inputs)
print(outputs)
```

```python
hidden_rep = outputs.last_hidden_state
cls_rep = outputs.pooler_output
print(hidden_rep)
print(cls_rep)
```

然后，我们可以用中文进行输入尝试

```python
sentence = "你好，我的朋友，我的名字叫李华。"
inputs = tokenizer(sentence, return_tensors="pt")
print(inputs)
```

```python
outputs = model(**inputs)
print(outputs)
```

```python
hidden_rep = outputs.last_hidden_state
cls_rep = outputs.pooler_output
```

```python
print(hidden_rep)
print(hidden_rep.shape)
```

```python
print(cls_rep)
print(cls_rep.shape)
```

`hidden_rep`包含句子中所有标记的特征，而`cls_head`包含`[CLS]`标记的特征，它持有这句话的总特征。以这种方式，可以用与其他BERT模型相同的方式来使用预训练的M-BERT模型。

### M-BERT单一的模型是可以实现跨语言的知识迁移
例如，你可以使用英文数据集进行训练，然后使用中文数据集进行测试，但是这样的微调仍然可以取得较好的效果。

`XNLI`数据集就是这样做的，它使用英语数据集进行训练，然后使用多种语言的数据集进行测试。

**跨语言知识迁移的原因主要如下**
- M-BERT模型的通用性并不取决于词汇的重叠度
- M-BERT模型的通用性取决于类型相似度（主谓宾的语法结构）
- M-BERT模型可以处理语码混用文本，但不能处理音译文本
