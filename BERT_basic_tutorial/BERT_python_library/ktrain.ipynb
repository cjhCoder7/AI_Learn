### ktrain完成情感分析任务

我们使用亚马逊数字音乐评论数据集获得完整的评论数据，也可以获得数据的子集。在这个练习中，我们使用包含数字音乐评论的数据子集。

```python
import os
os.environ['TF_USE_LEGACY_KERAS'] = "1"
```

```python
! pip install ktrain
import ktrain
from ktrain import text
import pandas as pd
```

```python
df = pd.read_json(r'reviews_Digital_Music_5.json',lines=True)
```

```python
df.head()
```

对数据集进行删减，只保留“评论文本”和“总分”两列

```python
df = df[['reviewText','overall']]
```

```python
df.head()
```

评分范围为从1到5。我们把这些评分用于情感分类，将`1~3`的分数映射到负面类，将`4~5`的分数映射到正面类

```python
sentiment = {1:'negative',2:'negative',3:'negative',
             4:'positive',5:'positive'}
```

```python
df['sentiment'] = df['overall'].map(sentiment)
```

```python
df.head()
```

再次对数据集进行删减，只保留“评论文本”和“情感”两列

```python
df = df[['reviewText','sentiment']]
```

```python
df.head()
```

#### 创建训练集和测试集。

如果数据在pandas DataFrame中，使用text_from_df函数创建训练集和测试集。如果数据是在一个文件夹中，使用text_from_folder函数创建训练集和测试集

```python
# 从数据框(df)中获取训练集和测试集，并对文本进行BERT风格的预处理
(x_train, y_train), (x_test, y_test), preproc = text.texts_from_df(
    train_df=df,                    # 使用的数据框, 这里是'df'，包含文本和标签列
    text_column='reviewText',       # 数据框中存储文本的列名，文本会被用作模型的输入
    label_columns=['sentiment'],    # 数据框中存储标签的列名，这里是 'sentiment' 列，表示情感（正面/负面）
    maxlen=100,                     # 设置输入文本的最大长度，超过的部分会被截断，不足的会进行填充
    max_features=100000,            # 最大的词汇表大小，表示可以使用的最多词汇数量（影响模型的词汇量）
    preprocess_mode='bert',         # 选择预处理模式，这里使用 'bert'，表示将文本转换为 BERT 所需的输入格式
    val_pct=0.1                     # 用于验证集的比例，这里是 0.1 表示 10% 的数据作为验证集
)
```

#### 训练

接下来，定义分类器。在这之前，我们先看一下ktrain库提供的分类器

```python
text.print_text_classifiers()
```

从上面的结果可以看到，ktrain库提供了多样化的分类器，包括**logistic回归分类器**、**双向GRU分类器**以及**BERT模型**。在这个练习中，我们使用BERT模型

基于 BERT 模型创建一个文本分类器

```python
model = text.text_classifier(
    name='bert',                    # 指定使用的模型是 'bert'，即使用预训练的 BERT 模型进行文本分类
    train_data=(x_train, y_train),  # 训练数据，包含预处理后的输入文本(x_train)和标签(y_train)
    preproc=preproc,                # 预处理器对象，包含文本的预处理方式及相关配置
    metrics=['accuracy']            # 指定评估指标为 'accuracy'，即模型性能将用准确率来衡量
)
```

通过 ktrain 库创建一个学习器（`learner`）

```python
learner = ktrain.get_learner(
    model=model,                     # 上一步创建的文本分类模型，即基于BERT的模型
    train_data=(x_train, y_train),    # 训练数据，包含输入数据和标签
    val_data=(x_test, y_test),        # 验证数据，用于在训练过程中评估模型的性能
    batch_size=32,                    # 每次训练的批量大小，表示每次更新模型时处理的样本数量
    use_multiprocessing=True          # 是否使用多进程来加速数据处理，True表示启用
)
```

使用fit_onecycle函数来训练模型

```python
learner.fit_onecycle(
    lr=2e-5,                     # 学习率，表示训练过程中调整模型权重的步长
    epochs=1,                     # 训练的轮数，即模型将遍历整个训练数据集的次数
    checkpoint_folder='output'    # 指定检查点保存的文件夹，用于存储训练过程中保存的模型
)
```

#### 预测

```python
predictor = ktrain.get_predictor(learner.model, preproc)
```

```python
predictor.predict('I loved the song')
```

```python
from ktrain import text
import os
```

将程序运行目录跳转到BBC文件夹

```python
os.chdir(os.getcwd() + '/bbc')
```

第一步是初始化索引目录，它被用来为所有文档编制索引。我们不需要手动创建任何新目录，只需把索引目录的名字传给`initialize_index`函数，就可以创建索引目录。

```python
# 创建一个简单的问答系统，并初始化一个索引文件夹，用于存储问答数据
text.SimpleQA.initialize_index('index')
```

在初始化索引之后，我们需要对文档进行索引。因为所有的文档在一个文件夹中，所以我们使用`index_from_folder`函数。`index_from_folder`函数接受参数`folder_path`和`index_dir`，以确定在哪里有所需要的文档和索引

```python
text.SimpleQA.index_from_folder(
    folder_path='entertainment',  # 数据存放的文件夹路径
    index_dir='index'             # 索引文件保存的文件夹
)
```

下一步是创建SimpleQA类的一个实例，如下所示。我们需要传递索引目录作为参数

```python
qa = text.SimpleQA('index')
```

```python
answers = qa.ask('who had a global hit with where is the love?')
```

```python
qa.display_answers(answers[:5])
```

尝试另一个问题

```python
answers = qa.ask('who win at mtv europe awards?')
```

```python
qa.display_answers(answers[:5])
```
